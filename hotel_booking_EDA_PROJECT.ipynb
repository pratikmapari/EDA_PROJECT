# Step2: Handling missing values.

As of we already checked that some columns have missing value now applying best approach to deal with that missing values

````python
df.isnull().sum().sort_values(ascending=False)[:4] # All column with missing value
````

The agent columns look like having agent number but in some cases that might be possible that people 
not booked hotel(resort) by any agent so in that case there are null values so one thing concluded 
that we can fill this null values by zero......

````python
df['agent'].fillna(0,inplace=True)#Replaced all the null value by zero 
print('Null values in the agent column==',df['agent'].isnull().sum())
````

````python
df['company'].fillna(0,inplace=True)#Replaced all the null value by zero 
print('Null values in the company column==',df['company'].isnull().sum())
````

This 'children' column has 0 value which means 0 children were present in group of customers who made that transaction.
So,'nan' values are the missing values due to error of recording data.

We will replace the null values under this column with mean value of children.

````python
df['children'].unique()
````

````python
df['children'].fillna(df['children'].mean(),inplace=True)
print('Replaced all the null value by mean')
df['children'].unique()
````

Next column with missing value is 'country'. This column represents the country of origin of customer.
Since, this column has datatype of string. We will replace the missing value with the mode of 'country' column.

````python
df['country'].fillna('other',inplace=True)
print('Replaced all the null value by other')
print('Total missing value ==',df['country'].isnull().sum())
````

````python
#checking for all null values are removed or not .so thre is no null value in the dataset
df.isnull().sum()
````
